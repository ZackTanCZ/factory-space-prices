```python
def listofMedian(df, var, price):
    medDF = df[[var,price]].groupby(by=[var]).median().sort_values(by=[price],ascending=False)
    orderList = medDF.index.tolist()
    return orderList
```

```python
FloorlvlPriceDF = summaryDF[['Floor Level', 'Price']]

display(FloorlvlPriceDF)

plt.figure(figsize=(5, 3))  
sns.boxplot(x = FloorlvlPriceDF['Price'], y = FloorlvlPriceDF['Floor Level'], 
            order = listofMedian(summaryDF, 'Floor Level', 'Price') ,showfliers = False )
plt.title('Fig 1.2a - Price by Floor Type')


plt.tight_layout()  # Adjust subplot spacing
plt.show()
```
